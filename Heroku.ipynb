```python
import sqlite3
import pandas as pd
from pandas import DataFrame

conn = sqlite3.connect('CovidCases.sqlite')
c = conn.cursor()


# Create table
c.execute('''CREATE TABLE CovidCases
             ([Date] date PRIMARY KEY, [DailyCases] integer, [vix] integer, [sp] integer, [xal] integer)''')

conn.commit()
```

```python
read_clients = pd.read_csv (r'assets/data/covidCasesAndSettles.csv')   
read_clients.to_sql('CovidCases', conn, if_exists='append', index = False)
```
